```python
# Importing necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Sample data for startups (can be replaced with actual data)
data = {
    'Startup': ['Getir', 'Papara', 'Yemeksepeti', 'Armut', 'KolayIK', 'Biseri', 'Trendbox', 'Neredekal', 'Killer Ideas', 'Evmanya'],
    'Funding ($ millions)': [500, 50, 100, 25, 40, 10, 60, 20, 15, 30],
    'Employees': [5000, 400, 1500, 200, 100, 50, 200, 300, 80, 150],
    'Tech Focus': [3, 4, 5, 2, 2, 1, 5, 4, 1, 3]  # 1: AI, 2: Blockchain, 3: IoT, 4: E-commerce, 5: Fintech
}

# Convert data into a DataFrame
df = pd.DataFrame(data)

# Feature scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df[['Funding ($ millions)', 'Employees', 'Tech Focus']])

# Apply KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_data)

# Calculate silhouette score for model evaluation
sil_score = silhouette_score(scaled_data, df['Cluster'])

# Plotting the clusters
plt.scatter(df['Funding ($ millions)'], df['Employees'], c=df['Cluster'], cmap='viridis')
plt.xlabel('Funding ($ millions)')
plt.ylabel('Employees')
plt.title('Clustering of Turkish Startups')
plt.show()

# Output the results
print(f"Silhouette Score: {sil_score}")
print("\nClustered Startups:\n", df[['Startup', 'Cluster']])
```
